```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>Import Data</h2>
</div>

```python
iris = datasets.load_iris()
```

```python
iris_data = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
iris_data
```

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>Build PCA Model</h2>
</div>

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(iris_data)
```

```python
## lets apply pca 
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1])
```

```python
## Apply Agglomerative Cluster
## To Construct dendrogram
import scipy.cluster.hierarchy as sc
plt.figure(figsize=(20, 17))
plt.title('Dendrogram')

# creat dendrogram
sc.dendrogram(sc.linkage(X_pca, method='ward'))
plt.xlabel('Sample Index')
plt.ylabel('Eucledian Distance')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, metric = 'euclidean', linkage = 'ward')
cluster.fit(X_pca)
```

```python
cluster.labels_
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = cluster.labels_)
```
