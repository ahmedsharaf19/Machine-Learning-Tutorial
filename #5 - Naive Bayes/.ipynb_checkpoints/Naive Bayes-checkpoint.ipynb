<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>Import Libraries</h2>
</div>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>Create Data</h2>
</div>

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0, random_state=50)
```

```python
sns.scatterplot(x = pd.DataFrame(X)[0], y = pd.DataFrame(X)[1], hue = y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>Linear SVC Model</h2>
</div>

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
## predicr
y_pred = model.predict(X_test)
```

```python
## Metrics Of Model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(f'Accuracy = {accuracy_score(y_test, y_pred)}')
print('#'*20)
print('Confusion Matrix : ')
print(confusion_matrix(y_test, y_pred))
print('#'*20)
print('Classification Report : ')
print(classification_report(y_test, y_pred))
```

```python
# lets create non lineat sperable data
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0)
sns.scatterplot(x = pd.DataFrame(X)[0], y = pd.DataFrame(X)[1], hue = y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(X_train, y_train)
## predicr
y_pred = model.predict(X_test)

## Metrics Of Model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(f'Accuracy = {accuracy_score(y_test, y_pred)}')
print('#'*20)
print('Confusion Matrix : ')
print(confusion_matrix(y_test, y_pred))
print('#'*20)
print('Classification Report : ')
print(classification_report(y_test, y_pred))
```

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>SVC Model Kernel RBF</h2>
</div>

```python
# lets try rbf kernel 
from sklearn.svm import SVC
model = SVC(kernel='rbf')
model.fit(X_train, y_train)
## predicr
y_pred = model.predict(X_test)

## Metrics Of Model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(f'Accuracy = {accuracy_score(y_test, y_pred)}')
print('#'*20)
print('Confusion Matrix : ')
print(confusion_matrix(y_test, y_pred))
print('#'*20)
print('Classification Report : ')
print(classification_report(y_test, y_pred))
```

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>SVC Model With Kernel Polynomial</h2>
</div>

```python
# lets try polynomial
from sklearn.svm import SVC
model = SVC(kernel='poly')
model.fit(X_train, y_train)
## predicr
y_pred = model.predict(X_test)

## Metrics Of Model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(f'Accuracy = {accuracy_score(y_test, y_pred)}')
print('#'*20)
print('Confusion Matrix : ')
print(confusion_matrix(y_test, y_pred))
print('#'*20)
print('Classification Report : ')
print(classification_report(y_test, y_pred))
```

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>SVC Model With Kernel Sigmoid</h2>
</div>

```python
# lets try sigmoid
from sklearn.svm import SVC
model = SVC(kernel='sigmoid')
model.fit(X_train, y_train)
## predicr
y_pred = model.predict(X_test)

## Metrics Of Model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(f'Accuracy = {accuracy_score(y_test, y_pred)}')
print('#'*20)
print('Confusion Matrix : ')
print(confusion_matrix(y_test, y_pred))
print('#'*20)
print('Classification Report : ')
print(classification_report(y_test, y_pred))
```

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>SVC with Hyperparamters Tuning</h2>
</div>

```python
from sklearn.model_selection import GridSearchCV
parameters = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf', 'linear', 'poly', 'sigmoid']
}

grid_model = GridSearchCV(SVC(), param_grid=parameters, cv = 5)
grid_model.fit(X_train, y_train)

## predict
y_pred = grid_model.predict(X_test)

## Metrics Of Model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(f'Accuracy = {accuracy_score(y_test, y_pred)}')
print('#'*20)
print('Confusion Matrix : ')
print(confusion_matrix(y_test, y_pred))
print('#'*20)
print('Classification Report : ')
print(classification_report(y_test, y_pred))
```
