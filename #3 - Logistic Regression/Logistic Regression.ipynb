```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>Create Data</h2>
</div>

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>Logistic Model</h2>
</div>

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
## predicr
y_pred = model.predict(X_test)
```

```python
## Metrics Of Model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(f'Accuracy = {accuracy_score(y_test, y_pred)}')
print('#'*20)
print('Confusion Matrix : ')
print(confusion_matrix(y_test, y_pred))
print('#'*20)
print('Classification Report : ')
print(classification_report(y_test, y_pred))
```

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>Logistic Model With GridSearch And Cross Validation</h2>
</div>

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
parameters = {
    'penalty': ['l1', 'l2', 'elasticnet'],
    'C': [100, 10, 1.0, 0.1, 0.01],
    'solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
}
grid = GridSearchCV(LogisticRegression(), param_grid=parameters, cv = StratifiedKFold(), n_jobs=-1)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
## Metrics Of Model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
y_pred = grid.predict(X_test)
print(f'Accuracy = {accuracy_score(y_test, y_pred)}')
print('#'*20)
print('Confusion Matrix : ')
print(confusion_matrix(y_test, y_pred))
print('#'*20)
print('Classification Report : ')
print(classification_report(y_test, y_pred))
```

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>Logistic Model With Multiclassification</h2>
</div>

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=10,n_informative=3 ,n_classes=3, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(multi_class='ovr')
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)
## Metrics Of Model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(f'Accuracy = {accuracy_score(y_test, y_pred)}')
print('#'*20)
print('Confusion Matrix : ')
print(confusion_matrix(y_test, y_pred))
print('#'*20)
print('Classification Report : ')
print(classification_report(y_test, y_pred))
```

```python
# Lets Try HyperParamter Tuning
from sklearn.model_selection import GridSearchCV, StratifiedKFold
parameters = {
    'penalty': ['l1', 'l2', 'elasticnet'],
    'C': [100, 10, 1.0, 0.1, 0.01],
    'solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
}
grid = GridSearchCV(LogisticRegression(multi_class='ovr'), param_grid=parameters, cv = StratifiedKFold(), n_jobs=-1)
grid.fit(X_train, y_train)

## Metrics Of Model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
y_pred = grid.predict(X_test)
print(f'Accuracy = {accuracy_score(y_test, y_pred)}')
print('#'*20)
print('Confusion Matrix : ')
print(confusion_matrix(y_test, y_pred))
print('#'*20)
print('Classification Report : ')
print(classification_report(y_test, y_pred))
```

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>ROC Curve</h2>
</div>

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
X, y = make_classification(n_samples=1000, n_classes=2, random_state=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
# Alwayes Predict Tho Output is 0
dummy_model_prob = [0 for _ in range(len(y_test))]
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# prediction based on probability
model_prob = model.predict_proba(X_test)
```

```python
# Focus on the positive output
model_prob = model_prob[:, 1]
```

```python
dummy_model_auc = roc_auc_score(y_test, dummy_model_prob)
model_auc = roc_auc_score(y_test, model_prob)
print(dummy_model_auc)
print(model_auc)
```

```python
dummy_fpr, dummy_tpr,dummy_threshold = roc_curve(y_test, dummy_model_prob)
model_fpr, model_tpr, model_threshold = roc_curve(y_test, model_prob)
```

```python
plt.plot(dummy_fpr, dummy_tpr, linestyle = '-', label = 'Dummy Model')
plt.plot(model_fpr, model_tpr, marker = '.', label = 'Logistic')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 50))

# Plotting ROC curves
ax.plot(dummy_fpr, dummy_tpr, linestyle='-', label='Dummy Model')
ax.plot(model_fpr, model_tpr, marker='.', label='Logistic')

# Annotating thresholds
for fpr, tpr, thresh in zip(model_fpr, model_tpr, model_threshold):
    ax.annotate(f'{thresh:.2f}', xy=(fpr, tpr), textcoords="offset points", xytext=(0, 10), ha='center')

# Adding labels and legend
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.legend()

plt.show()
```
