```python
corpus = """Hello Welcome,to Krish Naik's NLP Tutorials.
Please do watch the entire course! to become expert in NLP.
"""

print(corpus)
```

```python
from nltk.tokenize import sent_tokenize
```

```python
# Convert Paragraph -> Sentences
documents = sent_tokenize(corpus)
print(type(documents))
```

```python
for sentence in documents:
    print(sentence)
```

```python
# Lets Convert (Paragraphs, Sentences) -> Words
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
print('Word Tokenizer')
print('----------------------------')
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize  # علشان يعتبر ال punctiuation هى كمان Token
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer # full stop not split in seperate word but in if fullstop in last in corpus will split in sperate word
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
