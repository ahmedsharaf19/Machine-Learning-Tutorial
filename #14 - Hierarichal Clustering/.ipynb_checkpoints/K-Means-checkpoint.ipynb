```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>Import Data</h2>
</div>

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
cancer_dataset = pd.DataFrame(cancer_dataset['data'], columns = cancer_dataset['feature_names'])
cancer_dataset
```

<div style = 'background-color: #6482AD; text-align:center; padding: 10px; color:#F5EDED'>
    <h2>Build PCA Model</h2>
</div>

```python
# Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(cancer_dataset)
scaled_data = scaler.transform(cancer_dataset)
```

```python
# Apply PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2) # 2 is number of principal component analysis
data_pca = pca.fit_transform(scaled_data)
data_pca
```

```python
# Get Explained Variance
pca.explained_variance_, pca.explained_variance_ratio_
```

```python
# Get Precentage Of Captured Variance 
np.sum(pca.explained_variance_ratio_)
```

```python
# if we increase number of principal component 
pca_inc = PCA(n_components=5)
data_pca_inc = pca_inc.fit_transform(scaled_data)
np.sum(pca_inc.explained_variance_ratio_)
```

```python
# lets draw data after pca
plt.figure(figsize = (8, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c = load_breast_cancer()['target'])
plt.xlabel('Fisrt Principal Component')
plt.ylabel('Second Principal Component')
```
